```python
# Basic Q2


def Is_Number_in_Ordered_List(mylist, number):
    for i in mylist:
        if i==number:
            return True
    return False

print (Is_Number_in_Ordered_List ([1,2,3],5))

print (Is_Number_in_Ordered_List ([1,2,3],2))






```

```python
#Basic Q3


def Remove_Odd_Numbers_from_List(mylist):
    for i in mylist:
        if (i % 2 != 0) :
            mylist.remove(i)
    return(mylist)

print(Remove_Odd_Numbers_from_List([1,4,9,16,25,36,49,64,81,100]))







```
