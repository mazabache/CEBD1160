```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Comment this if the data visualisations doesn't work on your side
%matplotlib inline

plt.style.use('bmh')
```

```python
#Basic Q1 and Q2

df = pd.read_csv('train_titanic.csv')
df.info() 
print("This is the list of passengers in the Titanic last voyage")
print("Did not capture the age of all passengers. Only 204 cabin members (assume crew members). Two people did not embark")
```

```python
#Basic Q3
df.drop(df.index[1]) #This could be index 0 if question meant first row in the data frame
```

```python
#Basic Q4
df.drop('Embarked', axis=1) #Note: axis=1 denotes that we are referring to a column, not a row
```

```python
#Basic Q5

#DataFrame.dropna()
#Axis: Specifies to drop by row or column. 0 means row, 1 means column.
#How: Accepts one of two possible values: any or all. This will either drop an axis which is completely empty (all), or an axis with even just a single empty cell (any).
#Thresh: Here's an interesting one: thresh accepts an integer, and will drop an axis only if that number threshold of empty cells is breached.
#Subset: Accepts an array of which axis' to consider, as opposed to considering all by default.
#Inplace: If you haven't come across inplace yet, learn this now: changes will NOT be made to the DataFrame you're touching unless this is set to True. It's False by default.

df.dropna(axis=0, how='any', thresh=None, subset=None, inplace=False) #Note: axis=0 denotes that we are referring to a row, not a column
```

```python
#Basic Q6

plt.figure(figsize=(9, 8))
sns.distplot(df['Fare'], kde=False);
```

```python
#Basic Q7
plt.figure(figsize=(9, 8))
sns.countplot(x="Survived", data=df);
```

```python
#Basic Q8

df_num = df.select_dtypes(include = ['float64', 'int64'])
corr = df_num.drop('Survived', axis=1).corr() 

plt.figure(figsize=(12, 10))
sns.heatmap(corr[(corr >= 0.3) | (corr <= -0.3)], 
            cmap='viridis', vmax=1.0, vmin=-1.0, linewidths=0.1,
            annot=True, annot_kws={"size": 8}, square=True);
```

```python
#Basic Q9
print("Interpretation of Results:")
print("Stronger Correlation between Class and Price. It was negative because higher class, lower number")
print("Stronger Correlation between the attributes Number of Siblings/Spouses vs Number of Parents/Childred. This is because of families. When travelling with children higher probability that spouse is also travelling")
print("The lower the age the higher probability that person is travelling with Siblings")
print("Not too many young people traveled in higher classes")
```

```python
#Advanced Q1
df['DoubledFare'] = df['Fare'] * 2
df
```

```python
#Advanced Q2
df['Age'].fillna((df['Age'].mean()), inplace=True) # False does not modify data frame, True does
df
```

```python
#Advanced Q3
plt.figure(figsize=(10, 10))

#Set style of scatterplot
sns.set_context("notebook", font_scale=1.1)
sns.set_style("ticks")

# Create scatterplot of dataframe
sns.lmplot('Age', # Horizontal axis
           'Fare', # Vertical axis
           data=df, # Data source
           fit_reg=False, # Don't fix a regression line
 #          hue="b", # Set color
           scatter_kws={"marker": "D", "s": 50}) #  marker size

# Set title
plt.title('Scatter Plot')

# Set x-axis label
plt.xlabel('Age')

# Set y-axis label
plt.ylabel('Fare')
```
