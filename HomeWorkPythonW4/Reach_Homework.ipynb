```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os
import numpy as np

# Comment this if the data visualisations doesn't work on your side
%matplotlib inline

df = pd.read_csv('train_titanic.csv')

os.makedirs('plot_titanic', exist_ok=True)
os.chdir('.\plot_titanic')
```

```python

fig, axes = plt.subplots(1, 1, figsize=(10, 10))
sns.boxplot('Sex', 'Fare', data=df, palette='inferno')
sns.swarmplot('Sex', 'Fare', data=df, color='black')
plt.savefig('box_plot_with_whiskers.png')
```

```python
fig, axes = plt.subplots(1, 1, figsize=(10, 10))
sns.boxplot('Fare', 'Sex', orient="h", data=df, palette='pastel')
sns.swarmplot('Fare', 'Sex', orient="h",data=df, color='black')
plt.savefig('box_plot_with_whiskers.png')
```
